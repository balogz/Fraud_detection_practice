#Encoding the categorical data using LabelEncoder and OnehotEncoder

```python
le = LabelEncoder()
```

```python
fraud_test['gender']= le.fit_transform(fraud_test['gender'])
fraud_test['category']= le.fit_transform(fraud_test['category'])
```

```python
#get one-hot encoding for zip columns
OHE_income = pd.get_dummies(fraud_test['zip'])
#drop the income columns since it is now encoded
fraud_test = fraud_test.drop('zip', axis =1)
#join the one hot encoded to the main data
fraud_test = fraud_test.join(OHE_income)
fraud_test.head(1)
```

```python
fraud_test.head()
```

```python
X = fraud_test.drop(columns=['Unnamed: 0','merchant','first','last','street','city','state','job','is_fraud','city_pop','trans_num','trans_date_trans_time','dob'],axis=1)
Y = fraud_test['is_fraud']
```

```python
print(X)
```

```python
X.head()
```

```python
X.info()
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

#Implementing cross validations to select the best model

---



```python
models = [KNeighborsClassifier(3), SVC(gamma=2, C=1), DecisionTreeClassifier(max_depth=5),  RandomForestClassifier(n_estimators=20),MLPClassifier(alpha=1, max_iter=1000),LogisticRegression()]
```

```python
def cross_val():
  for model in models:
    cv_score = cross_val_score(model, X, Y, cv=5)
    mean_accuracy = sum(cv_score)/len(cv_score)
    mean_accuracy = mean_accuracy*100
    mean_accuracy = round(mean_accuracy, 2)

    print('Cross Validation accuracies for the',model,'=', cv_score)
    print('Acccuracy score of the ',model,'=',mean_accuracy,'%')
    print('---------------------------------------------------------------')
```

```python
cross_val()
```

#Train and test splitting

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
dt = DecisionTreeClassifier(max_depth=5) # Define classifier
dt.fit(X_train, Y_train) # Train model

# Make predictions
y_train_pred = dt.predict(X_train)
y_test_pred = dt.predict(X_test)

# Training set performance
dt_train_accuracy = accuracy_score(Y_train, y_train_pred) # Calculate Accuracy
#dt_train_mcc = matthews_corrcoef(Y_train, y_train_pred) # Calculate MCC
#dt_train_f1 = f1_score(Y_train, y_train_pred, average='weighted') # Calculate F1-score

# Test set performance
dt_test_accuracy = accuracy_score(Y_test, y_test_pred) # Calculate Accuracy
#dt_test_mcc = matthews_corrcoef(Y_test, y_test_pred) # Calculate MCC
#dt_test_f1 = f1_score(Y_test, y_test_pred, average='weighted') # Calculate F1-score

print('Model performance for Training set')
print('- Accuracy: %s' % dt_train_accuracy)
#print('- MCC: %s' % dt_train_mcc)
#print('- F1 score: %s' % dt_train_f1)
print('----------------------------------')
print('Model performance for Test set')
print('- Accuracy: %s' % dt_test_accuracy)
#print('- MCC: %s' % dt_test_mcc)
#print('- F1 score: %s' % dt_test_f1)
```
